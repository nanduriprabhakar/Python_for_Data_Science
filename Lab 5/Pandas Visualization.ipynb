## Plot Types

There are several plot types which are built-in to pandas. 
* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

There are 2 ways by which you can specify the plot:
>1) df.plot.area

or you can say

>2) df.plot(kind='area)

you can replace kind argument with any of the key plot types shown in the list above

### Area

```python
df2.head()
```

```python
df2.plot.area(alpha = 0.5)
#you can change alpah for transparancy of area plot
```

### Bar

```python
df2.plot(kind ='bar')
```

```python
df2.plot(kind ='bar',stacked = True)
```

### Histogram

```python
df1['A'].plot.hist(bins = 50)
```

```python
df1['A'].plot.hist(bins = 10)
```

### Line Plots

```python
df1.plot.line(x=df1.index, y='B', figsize = (12,5), lw=1)
```

```python
df1.plot.line(x=df1.index, y='B', figsize = (12,10), lw=6)
```

### Scatter plots 

```python
df1.plot.scatter(x='A' , y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C', cmap = 'coolwarm')
# s stands for size. s needs to be an array denoting sizes for different columns
# cmap 
```

```python
df1.plot.scatter(x='A', y ='B', s=df1['C']*100)
```

### Box Plots

```python
df2.plot.box()
#we can alos pass by= argument for groupby
```

### Hexagonal Bin Plot

hexagonal plots are useful for Bivariate data, alternative to scatterplot

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a', 'b'])
df.plot.hexbin(x='a', y='b', gridsize= 30, cmap = 'Oranges')
```

### Kernel Density Estimation Plot (KDE)

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
